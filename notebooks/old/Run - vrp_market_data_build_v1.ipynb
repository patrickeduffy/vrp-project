```python
from pathlib import Path
import re

script_path = Path(r"C:\Users\patri\vrp_project\notebooks\vrp_market_data_build_v1.py")

text = script_path.read_text(encoding="utf-8")

new_func = '''def parse_yyyymmdd(value: str) -> pd.Timestamp:
    try:
        parsed = pd.to_datetime(str(value), format="%Y%m%d", errors="raise")
        return pd.Timestamp(parsed).normalize()
    except Exception as exc:
        raise ValueError(f"Date must be YYYYMMDD. Got {value!r}") from exc

'''

pattern = r"def parse_yyyymmdd\(value: str\) -> pd\.Timestamp:\n.*?\n(?=def make_chunks)"

patched, n = re.subn(pattern, new_func, text, flags=re.DOTALL)

if n != 1:
    print(f"Patch count: {n}")
    print("Could not patch automatically. Showing nearby text:")
    idx = text.find("def parse_yyyymmdd")
    print(text[idx:idx+800])
    raise RuntimeError("Patch failed.")

script_path.write_text(patched, encoding="utf-8")

print("Patched parse_yyyymmdd successfully.")
```
